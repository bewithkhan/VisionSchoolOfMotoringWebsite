```python
import pandas as pd

# Load the survey data from the excel file
df = pd.read_excel("/content/Strategic Approaches to Customer Acquisition and sustainable Business Practices(1-49).xlsx")

# Display the first few rows of the dataframe
display(df.head())
```

```python
# Drop unnecessary columns
df_cleaned = df.drop(columns=['ID', 'Start time', 'Completion time', 'Email', 'Name', 'Last modified time',
                               'Please read the following statements and tick each box to confirm your agreement before starting the survey:',
                               'What is the purpose of this study?', 'Why have I been invited to take part?',
                               'What will I be asked to do?', 'Do I have to take part?',
                               'Will my data be confidential?', 'What will happen to the results of the study?',
                               'Are there any risks or benefits?', 'Who can I contact for more information?',
                               'Provide recommendations on how Interloop can enhance its sustainability strategies to better acquire and retain customers (optional):',
                               'Additional Comments (optional)'])

# Rename columns for better readability
df_cleaned.columns = ['Department', 'Experience', 'Role in Sustainability',
                      'Customer Perception of Sustainability', 'Sustainability in Purchasing Decisions',
                      'Eco-Labeling Effectiveness', 'Green Marketing Effectiveness',
                      'Challenges in Adopting Sustainability', 'Competitive Advantage of Sustainability',
                      'Impact of SDG Alignment', 'Role of Middle Management', 'Impact of Greenwashing',
                      'Importance of Transparency', 'Future of Sustainability in Textiles',
                      'Key Drivers for Sustainability Adoption', 'Barriers to Implementing Circular Economy',
                      'Role of Digitalization in Sustainability', 'Ethical Sourcing and Customer Acquisition',
                      'Impact of COVID-19 on Sustainability', 'Measuring ROI of Sustainability',
                      'Collaboration for Sustainability', 'Consumer Education on Sustainability',
                      'Government Role in Promoting Sustainability', 'Influence of Fast Fashion',
                      'Innovation in Sustainable Materials', 'Supply Chain Transparency',
                      'Employee Engagement in Sustainability', 'Long-Term Value of Sustainability']

# Display the first few rows of the cleaned dataframe
display(df_cleaned.head())
```

## Handle missing values

### Subtask:
Handle missing values in the `df_cleaned` dataframe.


**Reasoning**:
I will first identify the missing values in the `df_cleaned` dataframe. Then, I will iterate through each column, check if it has missing values, and fill them with the mode for categorical columns and the mean for numerical columns. Finally, I will verify that there are no more missing values.



```python
import numpy as np

# Check for missing values
print("Missing values before handling:")
display(df_cleaned.isnull().sum())

# Handle missing values
for column in df_cleaned.columns:
    if df_cleaned[column].isnull().any():
        if df_cleaned[column].dtype == 'object':
            df_cleaned[column].fillna(df_cleaned[column].mode()[0], inplace=True)
        else:
            df_cleaned[column].fillna(df_cleaned[column].mean(), inplace=True)

# Verify that there are no more missing values
print("\nMissing values after handling:")
display(df_cleaned.isnull().sum())
```

## Analyze categorical variables

### Subtask:
Analyze the distribution of categorical variables like 'Department', 'Experience', and 'Role in Sustainability'.


**Reasoning**:
Select the specified categorical columns and calculate the value counts for each to analyze their distribution.



```python
# Select categorical columns for analysis
categorical_cols = ['Department', 'Experience', 'Role in Sustainability']
df_categorical = df_cleaned[categorical_cols]

# Calculate and display value counts for each categorical column
print("Distribution of Department:")
display(df_categorical['Department'].value_counts())

print("\nDistribution of Experience:")
display(df_categorical['Experience'].value_counts())

print("\nDistribution of Role in Sustainability:")
display(df_categorical['Role in Sustainability'].value_counts())
```

## Analyze the relationship between sustainability and customer behavior

### Subtask:
Investigate the relationship between sustainability initiatives and customer purchasing decisions.


**Reasoning**:
To investigate the relationship between sustainability initiatives and customer purchasing decisions, I will first import the necessary visualization libraries, then create a box plot to visualize the distribution of 'Sustainability in Purchasing Decisions' for each 'Role in Sustainability' category, and finally generate and visualize a correlation matrix for the specified columns.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Role in Sustainability', y='Sustainability in Purchasing Decisions', data=df_cleaned)
plt.title('Sustainability in Purchasing Decisions by Role in Sustainability')
plt.xlabel('Role in Sustainability')
plt.ylabel('Sustainability in Purchasing Decisions')
plt.xticks(rotation=45)
plt.show()

# Generate a correlation matrix
correlation_matrix = df_cleaned[['Customer Perception of Sustainability',
                                 'Sustainability in Purchasing Decisions',
                                 'Eco-Labeling Effectiveness',
                                 'Green Marketing Effectiveness',
                                 'Impact of SDG Alignment']].corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Sustainability Factors and Purchasing Decisions')
plt.show()
```

## Analyze the effectiveness of green marketing

### Subtask:
Evaluate the effectiveness of different green marketing strategies based on the survey data.


**Reasoning**:
Create bar plots to visualize the distribution of responses for 'Eco-Labeling Effectiveness' and 'Green Marketing Effectiveness'.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot for Eco-Labeling Effectiveness
plt.figure(figsize=(8, 5))
sns.countplot(x='Eco-Labeling Effectiveness', data=df_cleaned, palette='viridis')
plt.title('Distribution of Responses for Eco-Labeling Effectiveness')
plt.xlabel('Eco-Labeling Effectiveness (1-5 Scale)')
plt.ylabel('Frequency')
plt.show()

# Create a bar plot for Green Marketing Effectiveness
plt.figure(figsize=(8, 5))
sns.countplot(x='Green Marketing Effectiveness', data=df_cleaned, palette='viridis')
plt.title('Distribution of Responses for Green Marketing Effectiveness')
plt.xlabel('Green Marketing Effectiveness (1-5 Scale)')
plt.ylabel('Frequency')
plt.show()

# Analyze the value counts for both columns
print("Value counts for Eco-Labeling Effectiveness:")
display(df_cleaned['Eco-Labeling Effectiveness'].value_counts().sort_index())

print("\nValue counts for Green Marketing Effectiveness:")
display(df_cleaned['Green Marketing Effectiveness'].value_counts().sort_index())
```

## Identify challenges and opportunities

### Subtask:
Identify the main challenges and opportunities for implementing sustainable practices in the textile industry.


**Reasoning**:
To identify the main challenges and opportunities, I will first analyze the 'Challenges in Adopting Sustainability' and 'Competitive Advantage of Sustainability' columns. I'll use `value_counts()` to get the frequency of each response and then create bar plots to visualize these distributions, which will make it easy to see the most cited challenges and benefits.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analyze challenges
challenges = df_cleaned['Challenges in Adopting Sustainability'].value_counts()
print("Challenges in Adopting Sustainability:")
display(challenges)

plt.figure(figsize=(10, 6))
sns.barplot(x=challenges.index, y=challenges.values, palette='Reds_d')
plt.title('Top Challenges in Adopting Sustainability')
plt.xlabel('Challenges')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Analyze competitive advantages
advantages = df_cleaned['Competitive Advantage of Sustainability'].value_counts()
print("\nCompetitive Advantage of Sustainability:")
display(advantages)

plt.figure(figsize=(10, 6))
sns.barplot(x=advantages.index, y=advantages.values, palette='Greens_d')
plt.title('Top Competitive Advantages of Sustainability')
plt.xlabel('Competitive Advantages')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Summary and Recommendations

### Key Findings

*   **Sustainability drives purchasing decisions:** There is a strong positive correlation between how customers perceive a company's sustainability efforts and their likelihood of making purchasing decisions based on sustainability. This is further supported by the fact that individuals more involved in sustainability initiatives are more likely to prioritize it in their purchasing decisions.
*   **Green marketing is effective:** Both eco-labeling and green marketing are perceived as highly effective strategies. A significant number of respondents rated both as highly effective (4 or 5 out of 5).
*   **Challenges and Opportunities:**
    *   **Top Challenges:** The most significant challenges to adopting sustainable practices are high implementation costs, lack of consumer awareness, and complex supply chains.
    *   **Top Opportunities:** The most significant competitive advantages of sustainability are enhanced brand reputation, increased customer loyalty, and innovation in products and processes.
*   **Demographics:** The majority of survey respondents are from the Marketing department, have 2-5 years of experience, and are moderately involved in sustainability initiatives.

### Recommendations

*   **Invest in consumer education:** Since a lack of consumer awareness is a major hurdle, companies should invest in targeted marketing and educational campaigns to inform consumers about the benefits of sustainable textiles. This will not only address a key challenge but also leverage the effectiveness of green marketing.
*   **Highlight sustainability efforts in branding:** Given that an enhanced brand reputation is a major competitive advantage, companies should prominently feature their sustainability efforts in their branding and marketing materials. This will help attract and retain customers who prioritize sustainability.
*   **Focus on innovation:** To address the challenge of high implementation costs, companies should focus on innovation in sustainable materials and processes. This can lead to cost reductions in the long run and also provide a competitive advantage.
*   **Engage middle management:** Since middle management plays a crucial role in implementing sustainability initiatives, it is important to engage them in the process. This can be done through training, incentives, and clear communication of sustainability goals.
*   **Improve supply chain transparency:** Given the complexity of supply chains, companies should work on improving transparency. This can be achieved through the use of technology, such as blockchain, and by working closely with suppliers.
